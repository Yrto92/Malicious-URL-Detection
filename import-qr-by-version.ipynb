```python
path = "/Users/yaeerk/Documents/NAYA/FinalProject/"
folder = f"/Users/yaeerk/Documents/NAYA/FinalProject/QRcode/"

version = 12 #65x65
```

```python
#https://www.kaggle.com/datasets/sid321axn/malicious-urls-dataset
df = pd.read_csv(path + "malicious_phish.csv")
```

```python
df.head()
```

```python
types = df.type.unique().tolist()
counter = {}
for t in types:
    counter[t] = 0
    
n = 90000
finish = 0
saved = 0

for index, row in df.iterrows():
    
    try:
        
        if(counter[row.type] >= n):
            continue
                    
        qr = qrcode.make(row.url, version = version, border = 0)
        png_path = folder + row.type + "//" + str(index) + '.png'
        qr.save(png_path)
                
        counter[row.type] += 1
        
        saved += 1
        if(saved % 1000 == 0):
            print(f"{saved} ... {index}/{len(df)}   {counter} / {n}")            
        
        if(counter[row.type] >= n):
            finish += 1
            if(finish >= len(types)):
                break
        
    except:
        print(f"unable to qr: {row.type}:::{row.url}")
                    
print(f"Done")
```
